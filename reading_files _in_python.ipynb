### reading the files from multiple directories
- when data is present in multiple directories, we use the 'glob' library to list the files.
- here, we split the same dataset year-wise into multiple files and stored them in multiple directories.

```python
# now lets import glob library
import glob 
```

```python
# list all the files in the folder
for directory in glob.glob(r'C:\Users\Solanki Srikanth\Downloads\reading_csv-200323-192251\reading_csv\datasets\multi-directory\*'):
    print(directory)
```

now each the folders 1985,1987,1997,1998,1999,2002,2004,2007 and 2009 contains CSV. We will iterate through each of them and concatente the files.

```python
# iterate through each of the files
for directory in glob.glob(r'C:\Users\Solanki Srikanth\Downloads\reading_csv-200323-192251\reading_csv\datasets\multi-directory\*'):
    for files in glob.glob(directory+ '\*'):
        print(files)
```

### now lets concatenete the files


```python
# list to store data frames
data_frame_list = []

for files in glob.glob(directory + '\*'):
    data_frame_list.append(pd.read_csv(files))
                           
                    
            
                           
        
```

```python
# concatenate the dataframes
final_data = pd.concat(data_frame_list)
final_data.head()
```

```python
# store the csv
final_data.to_csv('datasets/merged_sales.csv', index=False)
```
